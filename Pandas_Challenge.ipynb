# 1. Subset Rows by Row Number

We subset rows by row number when we want to extract specific rows from a DataFrame based on their position in the DataFrame. This allows us to focus on a subset of the data that meets our criteria. For example, in a farms dataset, we may want to extract the first 10 rows to analyze the initial farms.

To subset rows by row number, we can use the iloc method in Pandas. We specify the row numbers we want to extract and pass them as arguments to the iloc method. This will return a new DataFrame containing only the specified rows.

```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/agriculture_fjci5.csv')
df.head()
```

```python
# Subset rows by row number to extract the first 5 rows
subset_rows = df.iloc[:]

# Display the subset of rows
subset_rows
```

This can provide valuable insights into the revenue generated by each product.

To apply the assign() function, we first select the DataFrame and then use the method chaining technique to add new columns. We can also use assign() to update existing columns by reassigning them with new values based on certain conditions or calculations.

# 5. Create a Summary Table by Grouping and Applying a Custom Function

We can create a summary table by grouping our data based on specific columns and applying a custom function to aggregate the information. This allows us to condense large datasets into a more manageable form, providing us with valuable insights and trends. For example, we can group sales data by region and calculate the total revenue generated in each region using a custom function.

To apply this concept, we first identify the columns we want to group by, then define the custom function to perform the aggregation. Finally, we use the Pandas library in Python to group the data, apply the custom function, and create the summary table.



```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/agriculture_sv1ue.csv')
df.head()
```

```python
# Define a custom function to calculate total profit per Crop Type
def calculate_total_profit(group):
    return group['Profit'].sum()

# Group the data by 'Crop Type' and apply the custom function
summary_table = df.groupby('Crop Type').apply(calculate_total_profit).reset_index()

# Rename the columns to include a name for the total profit column
summary_table.columns = ['Crop Type', 'Total Profit']

# Display the summary table
summary_table
```

# 6. Splitting Columns based on Conditions

We split columns based on conditions when we want to divide our dataset into multiple parts according to specific criteria. This allows us to analyze and manipulate different subsets of data separately. For example, we may want to split a dataset of customer information into two groups based on their age: customers under 30 and customers 30 and above. This can help us tailor marketing strategies for each group more effectively.

To apply this concept in Pandas, we can use the 'loc' function along with conditional statements to create new DataFrames containing the subsets of data that meet our criteria.



```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/agriculture_c0n63.csv')
df.head()
```

```python
# Split the 'Yield per Acre' column into two new columns based on a condition
df['High_Yield'] = df.loc[df['Yield per Acre'] >= 50, 'Yield per Acre']
df['Low_Yield'] = df.loc[df['Yield per Acre'] < 50, 'Yield per Acre']

# Display the modified DataFrame with the new columns
df.head()
```

# 7. Extracting Date Components

We extract date components to break down a date into its individual parts such as year, month, day, etc. This helps us analyze and manipulate time-based data more effectively. For example, we can extract the year from a date column to analyze yearly trends in a dataset. In real-life, we often need to extract date components for tasks like creating time series analysis, calculating age from birthdates, or grouping data by specific time periods.

To extract date components in Pandas, we can use the dt accessor with functions like dt.year, dt.month, dt.day, etc. This allows us to access and work with specific date components easily.



```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/agriculture_rwzfy.csv')
df.head()
```

```python
# Convert 'Planting Date' and 'Harvest Date' columns to datetime format
df['Planting Date'] = pd.to_datetime(df['Planting Date'])
df['Harvest Date'] = pd.to_datetime(df['Harvest Date'])

# Extracting year, month, and day components from 'Planting Date'
df['Planting_Year'] = df['Planting Date'].dt.year
df['Planting_Month'] = df['Planting Date'].dt.month
df['Planting_Day'] = df['Planting Date'].dt.day

# Extracting year, month, and day components from 'Harvest Date'
df['Harvest_Year'] = df['Harvest Date'].dt.year
df['Harvest_Month'] = df['Harvest Date'].dt.month
df['Harvest_Day'] = df['Harvest Date'].dt.day

# Displaying the updated DataFrame with extracted date components
df.head()
```

# 8. Performing Data Aggregation with GroupBy


We perform data aggregation with GroupBy when we want to group our data based on one or more columns and then apply a function to each group. This allows us to summarize and analyze our data at a higher level. For example, we can group sales data by region and then calculate the total sales for each region. This helps us to understand the overall performance of different regions.

To apply data aggregation with GroupBy in Pandas, we first create a GroupBy object by specifying the column(s) we want to group by. We then apply an aggregation function such as sum(), mean(), or count() to the GroupBy object to calculate the desired metrics for each group.

```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/agriculture_izqfl.csv')
df.head()
```

```python
# Grouping the data by 'Region' and calculating the total 'Yield' for each region
grouped_data = df.groupby('Region')['Yield'].sum().reset_index()

# Displaying the aggregated data
grouped_data
```

# 9. Aggregating Multiple Functions Simultaneously with the agg() Function

We use the agg() function in Pandas to apply multiple functions simultaneously to our data. This allows us to efficiently calculate various statistics or transformations on our dataset in a single step. For example, we can calculate the sum, mean, and standard deviation of multiple columns all at once. In real-life, we often need to analyze data from different perspectives, and aggregating multiple functions simultaneously helps us gain a comprehensive understanding of the dataset.

To apply this concept, we simply pass a dictionary to the agg() function where the keys are column names and the values are the functions we want to apply. This enables us to perform complex data manipulations with ease and flexibility.



```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/farmers_l5yt6.csv')
df.head()
```

```python
# Aggregating multiple functions simultaneously using the agg() function
agg_result = df.agg({
    'Total Acres': ['sum', 'mean', 'std'],
    'Total Production': ['sum', 'mean', 'std'],
    'Revenue': ['sum', 'mean', 'std']
})

# Displaying the aggregated results
agg_result
```
